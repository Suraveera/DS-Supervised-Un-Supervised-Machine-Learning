#### Filling the null values (Mean)

```python
df['LB']=df['LB'].fillna(df['LB'].mean())
df['AC']=df['AC'].fillna(df['AC'].mean())
df['DS']=df['DS'].fillna(df['DS'].mean())
df['DP']=df['DP'].fillna(df['DP'].mean())
```

### Standardization Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler=StandardScaler()
```

```python
df2['LB']=std_scaler.fit_transform(df2[['LB']])
df2['ASTV']=std_scaler.fit_transform(df2[['ASTV']])
df2['MSTV']=std_scaler.fit_transform(df2[['MSTV']])
df2['ALTV']=std_scaler.fit_transform(df2[['ALTV']])
df2['Ds']=std_scaler.fit_transform(df2[['DS']])
```
