### Standardization

```python
standardize_columns1=df.select_dtypes(include=['int'])
standardize_columns1
```

```python
def standardize_columns1(int):
    mean = int.mean()
    std_dev = int.std()
    standardized_columns1 = (int - mean) / std_dev
    return standardized_columns1
```

```python
for int in numerical_columns1:
    df[int] = standardize_columns1(df[int])

df[numerical_columns1].head()
```

```python
standardize_columns2=df.select_dtypes(include=['float'])
standardize_columns2
```

```python
def standardize_columns2(float):
    mean = float.mean()
    std_dev = float.std()
    standardized_columns2 = (float - mean) / std_dev
    return standardized_columns2
```

```python
for float in numerical_columns2:
    df[float]=standardize_columns2(df[float])

df[numerical_columns2].head()
```

```python
standardize_columns3=df.select_dtypes(include=['object'])
standardize_columns3
```

### Converting Numerical Columns Into Dummy Variables

```python
# Converting numerical columns into dummy variables
dummy_df = pd.get_dummies(df[All_columns123])
# Concatenating the original DataFrame with the dummy variables
df_with_dummies = pd.concat([df.drop(All_columns123, axis=1), dummy_df], axis=1)
# Now, df_with_dummies contains the original DataFrame with numerical columns converted to dummy variables
print(df_with_dummies.head())
```
