##  What are the key hyperparameters in KNN?

###1.  n_neighbors (k): This is the most important hyperparameter. It determines the number of nearest neighbors to consider when making a prediction.  A smaller k can lead to overfitting (sensitive to noise), while a larger k can lead to underfitting (may smooth out important patterns).  The optimal value often depends on the dataset and problem.

###   2. weights:** This parameter determines how the neighbors contribute to the prediction.
###   'uniform':  All neighbors have equal weight.'distance':  Neighbors closer to the query point have a greater influence.  This can be useful when closer points are more likely to be relevant.  You can also define custom weight functions.

### 3. algorithm:** This parameter specifies the algorithm used to compute the nearest neighbors. Options include:auto': Attempts to automatically select the best algorithm based on the values passed to `fit` method.'ball_tree'**: Uses a ball tree to find nearest neighbors efficiently.'kd_tree'**: Uses a KD tree to find nearest neighbors efficiently.'brute'**:  Performs a brute-force search.  This is simple but can be slow for large datasets.

    

##  What distance metrics can be used in KNN?

### 'euclidean'**:  Euclidean distance is the straight-line distance between two points in n-dimensional space. It's the most common distance metric.
