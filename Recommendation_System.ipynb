## Feature Extraction

```python
anime['genre'] = anime['genre'].fillna(1)
```

```python
anime
```

```python
mean_rating = anime['rating'].mean()
anime['rating'] = anime['rating'].fillna(mean_rating)
```

```python
anime_genres = anime['genre'].str.get_dummies(sep=', ')
anime_genres
```

```python
features = ['rating', 'members', 'episodes']
features
```

## Recommendation System:

```python
anime_title = "Cowboy Bebop"
anime_index = anime[anime['name'] == anime_title].index[0]
anime_index
```

```python
anime_features = pd.concat([anime[features], anime_genres], axis=1)
anime_features
```

```python
for col in anime_features.select_dtypes(include=['object']).columns:
            le = LabelEncoder()
            anime_features[col] = le.fit_transform(anime_features[col])
```

```python
cosine_similarities = cosine_similarity(anime_features)
cosine_similarities
```

```python
similar_anime_indices = cosine_similarities[anime_index].argsort()[:-10:-1]
similar_anime_indices
```

```python
similar_anime_indices = similar_anime_indices[1:]
similar_anime_indices
```

```python
 recommendations = anime['name'].iloc[similar_anime_indices].tolist()
 recommendations
```
