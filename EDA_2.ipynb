# MinMaxScaler
## Preferred when:
## 1. Data is not normally distributed and the algorithm is sensitive to feature scaling (e.g., KNN, K-means)
## 2. You want to ensure features are within a specific range, such as 0 to 1.  This is helpful for algorithms that might misinterpret large-value features as more important.
## 3. Neural Networks often benefit as it helps to avoid vanishing/exploding gradients during training.
## Why: MinMaxScaler transforms data into a specified range (default is 0 to 1), making features directly comparable and potentially improving algorithm performance.

# StandardScaler
## Preferred when:
## 1. Data is approximately normally distributed. Many statistical methods assume normal distribution.
## 2. You want to remove the mean and scale to unit variance. This centers data around zero.  
## 3. Algorithms that are sensitive to outliers (e.g., linear and logistic regression, SVM).  It reduces the impact of outliers compared to MinMaxScaler.
## Why: StandardScaler standardizes features by subtracting the mean and dividing by the standard deviation, making them zero-centered and with unit variance. This helps algorithms that assume normally distributed data.



# . Encoding Techniques

```python
categorical_cols = df.select_dtypes(include=['object']).columns
cols_to_encode = [col for col in categorical_cols if df[col].nunique() < 5]
cols_to_encode
```

```python
df = pd.get_dummies(df, columns=cols_to_encode, drop_first=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = df.select_dtypes(include=['object']).columns
cols_to_encode_label = [col for col in categorical_cols if df[col].nunique() >= 5]
label_encoders = {}
for col in cols_to_encode_label:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
df
```

```python
df['total_capital'] = df['capital_gain'] + df['capital_loss']
df
```

```python
df['age_education_interaction'] = df['age'] * df['education_num']
df
```

```python
df['capital_gain_log'] = np.log1p(df['capital_gain'])
df
```

```python
df['capital_loss_log'] = np.log1p(df['capital_loss'])
df
```

```python
df.drop(['capital_gain', 'capital_loss'], axis=1, inplace=True)
df
```

# . Feature Engineering

```python
from sklearn.ensemble import IsolationForest
model = IsolationForest(contamination='auto', random_state=42)

model
```

```python
df['outlier'] = model.fit_predict(df.drop('income_ >50K', axis = 1))
df
```

```python
df_cleaned = df[df['outlier'] == 1].drop('outlier', axis=1)
df_cleaned
```
